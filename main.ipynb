Import Data-Viz libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Generate a plot depicting the relative number of people who were insured vs not insured
○ Make sure to document your findings

```python
def custom_label(pct, allvals):
    absolute = int(pct / 100 * sum(allvals))
    return f"{pct:.1f}%\n({absolute})"

# Pie
data = hospital_df['insurance'].value_counts()
print(data)
labels = ['Insured', 'Uninsured']

plt.pie(data, labels=labels, autopct=lambda pct: custom_label(pct, data))
plt.title('Insured vs Uninsured visitors')
plt.show()
```

Generate plots depicting the correlation (if any) between a given type of hospital visit and the health of individuals.
○ I.e. Do people with health category, x, visit the hospital in a given manner more, the same, or less often than people in
health category, y, typed people?
○ Make sure to document your findings


```python
hospital_df['hospital_visits'] = hospital_df['ovisits'] + hospital_df['novisits'] + hospital_df['emergency']
grouped_df = hospital_df.groupby('health').agg(avg_visits=pd.NamedAgg(column='hospital', aggfunc='mean'))

sns.barplot(data=grouped_df, x='health', y='avg_visits', color='green')

plt.title('Hospital visits per health category')

plt.show()
```

Is there any relationship between income and private insurance? Between income and Medicaid?
○ Make sure to document your findings

For this exercise, we are assuming insurace value of 'yes' is private insurance

```python
# Private insurance is yes/no
# Income is unique per row, float

hospital_df['income_quartile'] = pd.qcut(hospital_df['income'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

grouped_df = hospital_df.groupby('income_quartile').agg(num_private=pd.NamedAgg(column='insurance', aggfunc=lambda x : (x == 'yes').sum()),
                                                        num_uninsured=pd.NamedAgg(column='insurance', aggfunc=lambda x : (x == 'no').sum()))

print(grouped_df)

sns.catplot(data = hospital_df, x='insurance', y='income', kind='box')
```
